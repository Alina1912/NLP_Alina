```python
negative = []
import os 
import nltk
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize 
from nltk.tokenize import word_tokenize 
directory = "archive/dataset/neg"
for filename in os.listdir(directory):# пробегаемся циклом по файлам 
    f = os.path.join(directory, filename) 
    text = open(f, "r")
    for line in text: 
        sents = word_tokenize(line) # каждую строчку текста токенизируем
        for word in sents:
            if word.isalpha():
                slovo = word.lower()
                p = morph.parse(slovo)[0] #лемматизируем 
                negative.append(p.normal_form) #добавляем слова в начальной форме в список negative
```

```python
positive = []
directory_pos = "archive/dataset/pos"
for filename in os.listdir(directory_pos):# пробегаемся циклом по файлам 
    f_p = os.path.join(directory_pos, filename) 
    text_pos = open(f_p, "r")
    for line in text_pos: 
        sents = word_tokenize(line) # каждую строчку текста токенизируем
        for word in sents:
            if word.isalpha():
                slovo = word.lower()
                p = morph.parse(slovo)[0] #лемматизируем 
                positive.append(p.normal_form) #добавляем слова в начальной форме в список positive
```

```python
positive_new = []
for i in positive:
    if i not in negative:
        positive_new.append(i)
print(positive_new)
```

```python
negative_new = []
for i in negative:
    if i not in positive:
        negative_new.append(i)
```

```python

from collections import Counter
freqlist_pos = {}
freqlist_pos = Counter()
for i in positive_new:
    freqlist_pos[i] += 1
sorted_dict_pos = {}
sorted_keys = sorted(freqlist_pos, key=freqlist_pos.get)  
for w in sorted_keys:
    sorted_dict_pos[w] = freqlist_pos[w]
p_sorted_dict = {}
for key,value in sorted_dict_pos.items():
    if value >= 3:  
        p_sorted_dict[key] = value
print(p_sorted_dict)
```

```python
from collections import Counter
freqlist_neg = {}
freqlist_neg = Counter()
for i in negative_new:
    freqlist_neg[i] += 1
sorted_dict_neg = {}
sorted_keys = sorted(freqlist_neg, key=freqlist_neg.get)  
for w in sorted_keys:
    sorted_dict_neg[w] = freqlist_neg[w]
n_sorted_dict = {}
for key,value in sorted_dict_neg.items():
    if value >= 3:  
        n_sorted_dict[key] = value
print(n_sorted_dict)
```

```python
freq_lists_common = {}
freq_lists_common['pos'] = dict(p_sorted_dict)
freq_lists_common['neg'] = dict(n_sorted_dict)
print(freq_lists_common)
```

```python
def tone_detect(freq_lists_common, text):
    counts = Counter()
    for lang, freq_list in freq_lists_common.items():
        freq_list = Counter(freq_list)
        for word in nltk.word_tokenize(text):
            counts[lang] += int(freq_list[word] > 0)
    return counts.most_common()
```

```python
file = open('463893-80.txt', "r")
q = file.read()
tone_detect(freq_lists_common, q)
```

```python
from sklearn.metrics import accuracy_score
outcome = []  # сюда будем писать результаты
test_tone = [] # сюда будем писать исходную тональность
directory = "archive/dataset/test"
for filename in os.listdir(directory):# пробегаемся циклом по файлам 
    f = os.path.join(directory, filename) 
    text = open(f, "r")
    q = text.read()
    predicted_result = tone_detect(freq_lists_common, q)[0][0]
    outcome.append(predicted_result)
    test_tone.append(filename[:3])
print(outcome) 
print(test_tone)
print("Accuracy: %.4f" % accuracy_score(outcome, test_tone))
```
