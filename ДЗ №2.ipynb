<h3>В текстовом файле 'DZ2.txt' содержится текст с предложениями, в которых возникают трудные моменты для автоматического посттегинга.</h3>

Среди них:

1) продуктивные модели (слова с неизвестным словарю корнем, но образованные с помощью продуктивных аффиксов).

2) сложные слова, у которых вторая часть совпадает со словами или вторыми частями сложных слов в словаре Зализняка. Пример: ультразвуковой, сверхчеловек, полугодовой

3) слова с «неизвестными» корнями (в т. ч. имена собственные), не содержащие продуктивных аффиксов, для которых носители языка могут однозначно определить лемму и часть речи (по стандартным окончаниям русского языка и зная контексты, в которых они употреблены). Пример: кринжатина, френдята, краш, агриться, тюрбо 

4) нестандартные деепричастия, которые употребляются в языке, но признаются окказиональными или ненормативными, в связи с чем обычно отсутствуют в словарях) Пример: жрамши, пимши, переодемшись

5) аббревиатуры , которые система могла бы спутать с глаголами или словами других классов и ошибиться в определении леммы. Пример: ТГМУ, НИУ ВШЭ, ЧСВ.

Прогоняем текст через <b>Pymorphy2 </b> и создаем список pos-тэггов

```python
#!pip3 install pymorphy2
from pymorphy2 import MorphAnalyzer
import nltk
from nltk.tokenize import sent_tokenize 
from nltk.tokenize import word_tokenize
m = MorphAnalyzer()
spisok_pm = []
f = open('DZ2.txt', "r", encoding="utf-8")
text = f.read()
text = text.lower()
import string
spec_chars = string.punctuation + '«»\t—…’'
text = "".join([ch for ch in text if ch not in spec_chars])
for token in text.split():
    p = m.parse(token)[0]
    spisok_pm.append(p.tag.POS)
print(spisok_pm)
```

```python
nlp = spacy.load("ru_core_news_sm")
f = open('DZ2.txt', "r", encoding="utf-8")
text = f.read()
import string
spec_chars = string.punctuation + '«»\t—…’'
text = "".join([ch for ch in text if ch not in spec_chars])
doc = Doc(text)
doc = nlp(text)
spisok_s = []
# Выведем токены, леммы и теги
for i, s in enumerate(doc.sents):
    print("\n-- Sentence %d --" % i)
    for t in s:
        spisok_s.append(t.pos_)
        if 'SPACE' in spisok_s:
            spisok_s.remove('SPACE')
print(spisok_s)
print(len(spisok_s))
```

Т.к. в разных системах имена тегов и части речи могут отличаться, –  сводим это всё к единому стандарту с помощью к кода-конвертера и сравниваем с нашим размеченным руками эталоном.

VERB = VERB, AUX, INFN

NOUN = NOUN, NUMR

ADV = ADV, ADVB, PRED

SCONJ = SCONJ

```python

for index, value in enumerate(spisok_s):
    if value == 'AUX':
        spisok_s[index] = 'VERB'
    if value == 'ADP':
        spisok_s[index] = 'PREP'
    if value == 'DET':
        spisok_s[index] = 'PRON'
    if value == 'SCONJ':
        spisok_s[index] = 'CONJ'
    if value == 'CCONJ':
        spisok_s[index] = 'CONJ'
print(len(spisok_s))
```
